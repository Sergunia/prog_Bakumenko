```python
#сделать с перемножением матриц
import numpy as np
import time
import matplotlib.pyplot as plt

def python(size_of_vec):
    start = time.time()
    X = range(size_of_vec)
    Y = range(size_of_vec)
    Z = [2*X[i]**2 + 4*Y[i] for i in range(len(X)) ]
    return time.time() - start

def numpy(size_of_vec):
    start = time.time()
    X = np.arange(size_of_vec)
    Y = np.arange(size_of_vec)
    Z = 2*X**2 + 4*Y
    return time.time() - start

def date():
    X = []
    Y = []
    y = 0
    for i in range(0, 1000, 1):
        for f in range(0, 20):
            y += (python(i) - numpy(i))        
        X.append(i)
        Y.append(y/20)
        y = 0
    return plt.plot(X, Y)

date()
plt.show()
```

```python
#Создать случаую квадратную матрицу случайного размера от 10 до 100.
#Найти максимум и сумму элементов.
#Поделить каждый элемент на максимум.
#Отнять от каждой строки матрицы среднее по строке
#Заменить максимальное значение на -1.
import random
size = random.randint(10,100)
a = np.random.random((size,size))
print(a, '\n')

print(np.sum(a))
max_ = np.max(a)
print(max_)

a /= max_
print(a)

midd = np.mean(a, axis = 1, keepdims = True)
a -= midd

print(a)
```
